**[7]**
Hasil dari*df.isnull().sum()* menunjukkan bahwa tidak ada nilai kosong (missing values) dalam dataset. Setiap kolom memiliki jumlah 0 yang berarti semua data sudah terisi dengan nilai yang lengkap.

Namun, meskipun tidak ada nilai kosong, terdapat kolom TotalCharges yang seharusnya berupa angka tetapi mungkin terbaca sebagai teks (object).

```python
departemen_biaya = df.groupby('Contract')['MonthlyCharges'].mean()
print(departemen_biaya)
```

**[8]** Hasil analisis menunjukkan bahwa pelanggan dengan kontrak bulanan memiliki rata-rata biaya bulanan tertinggi, yaitu sekitar 66.40. Sementara itu, pelanggan dengan kontrak One year membayar rata-rata 65.05, sedikit lebih rendah dibandingkan kontrak bulanan. Pelanggan dengan kontrak Two year membayar paling sedikit, dengan rata-rata biaya bulanan 60.77.

Artinya semakin lama durasi kontrak, maka rata-rata biaya bulanannya cenderung lebih rendah.Untuk mendapatkan biaya lebih murah pelanggan mungkin memilih kontrak tahunan.

```python
tenure_level = df.groupby('tenure')['MonthlyCharges'].mean()
tenure_level
```

**[9]** Hasil *df.groupby('tenure')['MonthlyCharges'].mean()* menunjukkan rata-rata biaya bulanan *(MonthlyCharges)* berdasarkan lama berlangganan *(tenure)* dalam bulan. Data ini memperlihatkan bahwa semakin lama pelanggan bertahan, rata-rata biaya bulanannya cenderung meningkat.

Contohnya pelanggan yang baru berlangganan 0 bulan memiliki rata-rata biaya sekitar 41.42, sementara pelanggan dengan masa berlangganan 72 bulan membayar rata-rata 80.70 per bulan. Artinya pelanggan yang bertahan lebih lama cenderung memilih layanan yang lebih mahal atau punya fitur tambahan dibandingkan pelanggan baru.

```python
import matplotlib.pyplot as plt
internet_service_distribution = df['InternetService'].value_counts(normalize=True) * 100

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(internet_service_distribution, labels=internet_service_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Internet Service')
plt.axis('equal')
plt.show()
```

**[10]** Berdasarkan data, mayoritas pelanggan menggunakan Fiber Optic (44%), diikuti oleh DSL (34.4%), sementara 21.7% pelanggan tidak menggunakan layanan internet.

**[Kesimpulan]**

Dari keseluruhan data, dapat disimpulkan bahwa pelanggan ternyata memiliki berbagai karakteristik dan preferensi layanan. Mayoritas pelanggan memilih kontrak bulanan/Month to Month yang cenderung lebih  fleksibel dibandingkan kontrak panjang tahunan.

Kemudian, Fiber Optic adalah layanan internet yang paling banyak digunakan, tetapi masih ada pelanggan yang tidak berlangganan internet. Rata-rata pelanggan membayar sekitar 64.76 per bulan dengan variasi yang cukup besar tergantung pada layanan yang digunakan dan lama berlangganan.

Selain itu, pelanggan dengan tenure lebih lama cenderung membayar biaya bulanan lebih tinggi yang mungkin menunjukkan peningkatan layanan atau paket tambahan.
